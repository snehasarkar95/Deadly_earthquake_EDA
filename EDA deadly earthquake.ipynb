```python
plt.figure(figsize=(15,15)) 
plt.pie(df['MMI'].value_counts(), labels=df['MMI'].value_counts().index, autopct='%1.1f%%', textprops={ 'fontsize': 20,
                                           'color': 'black',
                                           'weight': 'bold',
                                           'family': 'serif' }) 
hfont = {'fontname':'serif', 'weight': 'bold'}
plt.title('MMI', size=18, **hfont) 
plt.show()
```
